```python
# Import libraries

from bs4 import BeautifulSoup
import pandas as pd
import requests
import smtplib
import time
import datetime
```

```python
url = 'https://www.vlr.gg/event/stats/1657/valorant-champions-2023?exclude=&min_rounds=0&agent=all'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
# If you want to pull multiple tables from the same website with 
# similar tags reference 4:55 (Scraping Data from a Real Website)

soup.find('table')
```

```python
table = soup.find('table')
```

```python
headers = table.find_all('th')
```

```python
headers_list = [header.text.strip() for header in headers]

print(headers_list)
```

```python
df = pd.DataFrame(columns = headers_list)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print(individual_row_data)
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df.to_excel('ValProData.xlsx', index = False)
```
